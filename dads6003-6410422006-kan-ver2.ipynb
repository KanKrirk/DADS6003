```python
train=df_train.fillna(df_train.mean())
df_train.describe()
```

```python
df_train, test = train_test_split(train, test_size=0.2,random_state=42)
```

Imbalence(ROSE)


```python
y_0=df_train['y'][df_train['y']==0].count()
y_1=df_train['y'][df_train['y']==1].count()
print(y_0,y_1)
```

```python
train_pos = df_train[df_train['y']==1]
train_neg = df_train[df_train['y']==0]
```

```python
train_neg = train_neg.sample(int(df_train['y'].count()/2), random_state=42)
```

```python
train_pos.sample(int(df_train['y'].count()/2-(train_pos['y'].count())),random_state=42)
```

```python
train_pos_tmp=train_pos.sample(int(df_train['y'].count()/2-(train_pos['y'].count())),random_state=42)
train_pos=pd.concat([train_pos,train_pos_tmp])
train_pos
```

```python
train_pos.reset_index(drop=True)
train_pos.sort_index()
```

```python
new_train = pd.concat([train_neg, train_pos]).sort_index()
new_train['y'].count()
```

```python
hist = new_train.hist(bins=10)
```

```python
x_train=new_train.iloc[:,0:-1]
x_train
```

```python
y_train=new_train.iloc[:,-1]
y_train
```
